### Encoding 

#### Label Encoding
Colocamos labels de 0 a n em todos registros na ordem alfabética de uma variável categórica


```python
colunasEncoding = ['provedorInternet', 'tipoContrato', 'formaPagamento']
dataset_encode = dataset.drop(colunasEncoding, axis=1)
dataset_encode.columns
```

```python
encode = {'Sem Informação':2, 
          'Sem serviço de telefone':0,
          'Sem serviço de internet':0,
          'Feminino':1,
          'Masculino':0,
          'Não':0,
          'Sim':1
}
dataset_encode = dataset_encode.replace(encode)
dataset_encode.head()
```

```python
## buscando campos |
for col in dataset_encode.columns:
    null = dataset_encode[col].isnull().sum()
    

    print (f' A coluna {col} - Tem {null} registros null')
```

#### One-Hot Encoding 
Criamos n novas colunas dependendo dos valores unicos dentro de uma variável

```python
encoder = OneHotEncoder(dtype=int)

colunas_restantes = encoder.fit_transform(dataset[colunasEncoding]).toarray()

dataset_encode = pd.concat([dataset_encode, pd.DataFrame(colunas_restantes, columns=encoder.get_feature_names(colunasEncoding))], axis=1)
dataset_encode.head()
                                      
```

```python
## buscando campos |
for col in dataset_encode.columns:
    null = dataset_encode[col].isnull().sum()
    

    print (f' A coluna {col} - Tem {null} registros null')
```

#### Exportando o encode

## Verificar o balanceamento da Variável target

```python
dataset_encode.head()
```

Classes muito desbalanceadas = categorias com frequências muito diferentes.

Quando treinamos um modelo de classificação com a variável desbalanceada a classe dominante  vai se sobreassair em relação aos de menor frequência. 

GERALMENTE, nas bases temos a variável alvo desbalanceada (spoiler: o que é o nosso caso), e o que temos que prever é a menor frequência. 


#### Como podemos corrigir?

##### Undersampling 
Mantemos os dados da classe com menor frequência e diminuimos a quantidade na classe de maior frequência. 

- Vantagens: diminui o armazenamento de dados e o tempo de execução de códigos.
- Desvantagens: Acabamos excluindo registros que poderiam ser úteis para alimentar o ML. 

A técnica mais utilizada é a Near Miss que diminui aleatoriamente a quantidade de valores da classe maior, utilizando a menor distância média dos Kneighbors para reduzur a perda de informações

##### Oversampling 

Mantemos os dados da classe com maior frequência e aumentamos a quantidade na classe de menor frequência. 

- Vantagens: nenhuma informação dos registros que possuia é perdida, o que ajuda a alimentar o ML
- Desvantagens: Armazenamento e processamento crescem bastante, porém o modelo pode ficar 'Viciado', ou seja, vai prever muito bem nos dados de treinamento mas não também novos dados. 

A técnica mais utilizada é o SMOTE onde sintetizamos novas informações com base nas já existentes, esses dados sintéticos são próximos aos dados reais, porém não identicos.


#### Distribuição da Variável Target - desbalanceada

#### Aplicando o SMOTE (Oversampling)

Realizei a instalação de acordo com o tutorial
https://imbalanced-learn.org/stable/install.html#getting-started

- comando para instalar conda install -c conda-forge imbalanced-learn
 

```python
a =  dataset_encode.drop(['cancelouPlano'], axis=1)
b = dataset_encode['cancelouPlano'] 
```

```python
smote = SMOTE(random_state=42)
a_res, b_res = smote.fit_resample(a,b)

dataset = pd.concat([pd.DataFrame(a_res), pd.DataFrame(b_res)], axis=1)
```
